### <font color='red'>__Question I__</font> 
Create a figure sized 5 inches height and 10 inches width, composed of 3 subplots (a 1 by 3 grid)
- First subplot showing $y(t)=sin(\frac{1}{t})$ for $\frac{\pi}{5}\leq t \leq 5\pi$, line color green and line type dash-dot.
- Second one showing a tangent function from $\frac{\pi}{20}$ to $\frac{\pi}{5}$, line color black, hexagonal markers, dashed line style, and line thickness of 4.
- Third one showing $y(t)=1.5e^{-0.2t}\sin(5t+\frac{\pi}{4})$, plotted in blue color, for the range t=0 to t=10.

Make sure all the plots have grid lines, axes have labels, and there is a super title for the figure. Add code lines to save the figure with dpi=300 and to download the figure from the notebook. (You do __not__ need to submit the downloaded figure as part of your submission.)



```python
# Your code goes here
import matplotlib.pyplot as plt
import numpy as np

fig1, axs = plt.subplots(nrows=1,ncols=3,figsize=(10,5))
fig1.suptitle("Subplot Examples")

# First plot
x1 = np.linspace(np.pi/5, 5*np.pi, 25)
y1 = np.sin(1/x1)
axs[0].plot(x1,y1,'g-.')
axs[0].grid(b=True)
axs[0].set_xlabel('X1')
axs[0].set_ylabel('Y')

# Second plot
x2 = np.linspace(np.pi/20, np.pi/5, 25)
y2 = np.tan(x2)
axs[1].plot(x2,y2,'kh--',linewidth=4)
axs[1].grid(b=True)
axs[1].set_xlabel('X2')

# Third plot
x3 = np.linspace(0, 10, 100)
y3 = (1.5)*(np.exp(-.2*x3))*(np.sin(5*x3 + np.pi/4))
axs[2].plot(x3,y3,'b')
axs[2].grid(b=True)
axs[2].set_xlabel('X3')

plt.show()
fig1.savefig("InClassQ1Fig1.jpg", dpi=300)
```

### <font color='red'>__Question II__</font>
Using the UK energy generation distribution we used in the class:

- Plot a scatter plot showing the hydropower generation in GWh over the years in Scotland region.
Make sure you define the figure size, merkersize, marker color, and put labels for axes and put a title for your plot.

```python
# Your code goes here
import pandas as pd

# Import UK energy data from github
url = 'https://raw.githubusercontent.com/MasoudMiM/ME_364/main/UK_Renewable_Energy/UKEnergy.csv'
energy_data = pd.read_csv(url)

# Pull relevant info
#   The other file had commas in the numbers, but this file had tenfold higher values
#   Just going to accept those values as they are for now
scot_hydro_data = energy_data[energy_data['Region']=='Scotland'][['Year','Hydro']]

# Create scatterplot
fig2 = plt.figure(figsize=(8,6))
scot_hydro_data.plot.scatter(x='Year', y='Hydro', c='b', s=50, grid=True)
plt.title("Yearly Hydro Power Generation in Scotland")
plt.xlabel("Year")
plt.ylabel('Hydro (GWh)')
plt.show()
```

- Plot a bar plot representing solar power generation in GWh in different regions in the UK. Make sure you define the figure size, all the bars are in the same color, and the axes are labeled.

```python
# Your code goes here

# Solar power generation of each region varies by year
# I am assuming the question refers to the most recent year, 2015

# Get relevant info
solar_data = energy_data[energy_data['Year']==2015][['Region','Solar PV']]

solar_data

# Create bar graph
fig3 = plt.figure(figsize=(8,6))
solar_data.plot.bar(x='Region', y='Solar PV', color='orange')
plt.title('2015 UK Solar Power Generation by Region')
plt.ylabel('Solar Power (GWh)')
plt.show()
```
